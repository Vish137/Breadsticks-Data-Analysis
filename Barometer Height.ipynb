```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
```

```python
file = "Pressure Data.csv"
df = pd.read_csv(file)
s = df.to_numpy()
t, p = s.T
p = p[::-1] # reverse the order of array to show pressure from bottom to top
p = p*100 # convert hPa into Pa
```

```python
# Plot pressure v time
plt.plot(t,p,'-b',label="Pressure (Pa)")
plt.ylabel("Pressure (Pa)"); plt.xlabel("Time (s)")
plt.title("Pressure in moving elevator")
plt.grid()
plt.style.use(['science', 'notebook'])
```

```python
# Pressure differential
p_diff = p[0]-p[-1]

# Earth constants
g = 9.81 #m.s^2
rho = 1.23 #kg/m^3

# Height calculation
h = (g*rho)**(-1)*p_diff
print("Barometer height: ", round(h,2), "m")
```

```python
# Error calculation
h_stair = 15 # we need to actually measure the stair heights
num_stairs_floor = 22
floors = 9
height = (floors-1)*num_stairs_floor*h_stair
print('True height', height/100, 'm')
print('Accelerometer height', round(h,2),'m')
Error = abs((height/100) - h)/(height/100)
Error=Error*100
print('Error', round(Error,2), '%')
```
